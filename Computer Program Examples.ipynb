## 第一章　問題解決與運算思維

運算思維的四個核心面向：**問題分解、模式辨識、抽象化、演算法思考**。  
本章用一個「生產線良率分析」問題，示範如何用這四個思維拆解並解決問題。

### 情境：工廠有三條生產線，需要分析本週每條線的良率，並找出低於目標的生產線。

### 步驟 1　問題分解（Decomposition）

把「分析三條生產線的良率」這個大問題，拆解成幾個小任務：
1. 計算每條線的良率
2. 判斷每條線是否達標（目標：良率 ≥ 95%）
3. 列出未達標的生產線
4. 計算全廠平均良率

每個小任務都可以獨立撰寫，最後組合起來就是完整的分析。

```python
# ── 步驟 1：準備資料（每條生產線的總產量與不良品數）──

# 生產線名稱
lines = ["A線", "B線", "C線"]

# 各線本週總產量（件）
total = [500, 320, 410]

# 各線本週不良品數（件）
defects = [12, 8, 31]

print("資料準備完成")
print(f"生產線：{lines}")
print(f"總產量：{total}")
print(f"不良品：{defects}")
```

### 步驟 2　模式辨識（Pattern Recognition）

觀察三條線的計算方式完全相同：`良率 = (總產量 - 不良品數) / 總產量 × 100`  
這個重複的計算模式，可以用**迴圈**來統一處理。

```python
# ── 步驟 2：辨識出重複模式 → 用迴圈計算每條線的良率 ──

yield_rates = []  # 用來儲存每條線的良率

for i in range(3):  # i 依序為 0, 1, 2
    good = total[i] - defects[i]  # 良品數
    rate = good / total[i] * 100  # 良率（%）
    yield_rates.append(rate)  # 加入清單
    print(f"{lines[i]}：良品 {good} 件，良率 {rate:.1f}%")
```

### 步驟 3　抽象化（Abstraction）

「計算良率」這個動作可以包裝成一個**函數**，  
呼叫時只需給「總產量」和「不良品數」，不需在意內部細節。  
這就是抽象化——隱藏實作細節，暴露清楚的介面。

```python
# ── 步驟 3：將重複邏輯抽象化為函數 ──


def 計算良率(總產量, 不良品數):
    """輸入總產量與不良品數，回傳良率百分比"""
    良品數 = 總產量 - 不良品數
    良率 = 良品數 / 總產量 * 100
    return 良率


# 函數抽象化後，呼叫非常直觀
print(f"A線良率：{計算良率(500, 12):.1f}%")
print(f"B線良率：{計算良率(320,  8):.1f}%")
print(f"C線良率：{計算良率(410, 31):.1f}%")
```

### 步驟 4　演算法思考（Algorithmic Thinking）

將整個分析流程，寫成**清楚的步驟序列**：
1. 計算每條線的良率
2. 與目標值比較，判斷是否達標
3. 列出未達標的生產線
4. 計算並顯示全廠平均良率

```python
# ── 步驟 4：完整演算法——從資料到結論 ──

目標良率 = 95.0  # 目標：良率需達 95%
未達標清單 = []

print("=" * 35)
print("本週生產線良率分析報告")
print("=" * 35)

# 步驟 1 & 2：計算良率並判斷達標
for i in range(3):
    rate = 計算良率(total[i], defects[i])
    達標 = rate >= 目標良率
    狀態 = "✓ 達標" if 達標 else "✗ 未達標"
    print(f"{lines[i]}：{rate:.1f}%  {狀態}")
    if not 達標:
        未達標清單.append(lines[i])

# 步驟 3：列出未達標生產線
print("-" * 35)
if 未達標清單:
    print(f"需要改善的生產線：{未達標清單}")
else:
    print("所有生產線均達標！")

# 步驟 4：全廠平均良率
全廠良率 = sum(yield_rates) / len(yield_rates)
print(f"全廠平均良率：{全廠良率:.1f}%")
print("=" * 35)
```

### 3-2　for 迴圈

```python
# 用 for 迴圈處理一週的生產紀錄

星期 = ["週一", "週二", "週三", "週四", "週五"]
每日產量 = [480, 512, 498, 467, 521]
每日不良 = [12, 8, 15, 22, 9]

目標良率 = 95.0
總產量 = 0
低於目標天數 = 0

print(f"{'日期':4}  {'產量':>6}  {'不良':>4}  {'良率':>7}  {'狀態'}")
print("-" * 40)

for i in range(5):  # i 從 0 到 4
    良率 = (每日產量[i] - 每日不良[i]) / 每日產量[i] * 100
    總產量 += 每日產量[i]  # 累計總產量
    狀態 = "✓" if 良率 >= 目標良率 else "✗"
    if 良率 < 目標良率:
        低於目標天數 += 1
    print(f"{星期[i]:4}  {每日產量[i]:>6}  {每日不良[i]:>4}  {良率:>6.1f}%  {狀態}")

print("-" * 40)
print(f"週總產量：{總產量} 件")
print(f"低於目標天數：{低於目標天數} 天")
```

### 3-3　while 迴圈

```python
# 模擬生產線運作：生產到達配額為止

配額 = 100  # 今日生產目標
已生產 = 0
批次大小 = 12  # 每次加工一個批次
批次編號 = 0

print(f"今日配額：{配額} 件")
print("-" * 30)

while 已生產 < 配額:
    批次編號 += 1
    # 避免超產：本批次只生產到剛好達標
    本批產量 = min(批次大小, 配額 - 已生產)
    已生產 += 本批產量
    print(f"批次 {批次編號:02d}：生產 {本批產量} 件，累計 {已生產} 件")

print("-" * 30)
print(f"配額完成！共執行 {批次編號} 個批次")
```

### 3-4　break 與 continue

```python
# break：偵測到嚴重異常立即停止

溫度資料 = [22.1, 23.4, 21.8, None, 22.9, 31.5, 23.1]
#                                ↑            ↑
#                            感測器離線    緊急高溫

緊急上限 = 30.0

print("開始逐筆分析感測器資料...")

for i, 溫度 in enumerate(溫度資料):
    # continue：跳過空值（感測器離線）
    if 溫度 is None:
        print(f"  第 {i+1} 筆：感測器離線，略過")
        continue

    # break：溫度超過緊急上限，立即中止
    if 溫度 > 緊急上限:
        print(f"  第 {i+1} 筆：{溫度}°C ── 🔴 緊急高溫！停止分析，執行停機程序")
        break

    print(f"  第 {i+1} 筆：{溫度}°C ── 正常")

print("分析結束")
```

## 第四章　函數與模組化

### 4-1　定義與呼叫函數

```python
# 函數定義：計算良率
def 計算良率(總產量, 不良品數):
    """輸入總產量與不良品數，回傳良率百分比（0~100）"""
    if 總產量 == 0:
        return 0  # 避免除以零
    良品數 = 總產量 - 不良品數
    return 良品數 / 總產量 * 100


# 函數呼叫：同一個函數可以重複使用
A線良率 = 計算良率(500, 12)
B線良率 = 計算良率(320, 8)
C線良率 = 計算良率(410, 31)

print(f"A線良率：{A線良率:.1f}%")
print(f"B線良率：{B線良率:.1f}%")
print(f"C線良率：{C線良率:.1f}%")
```

```python
# 函數可以有預設參數值
def 判斷達標(良率, 目標=95.0):
    """判斷良率是否達到目標，目標預設為 95%"""
    if 良率 >= 目標:
        return True, "達標 ✓"
    else:
        差距 = 目標 - 良率
        return False, f"未達標 ✗（差 {差距:.1f}%）"


# 使用預設目標 95%
達標, 訊息 = 判斷達標(A線良率)
print(f"A線：{訊息}")

達標, 訊息 = 判斷達標(C線良率)
print(f"C線：{訊息}")

# 指定不同目標
達標, 訊息 = 判斷達標(C線良率, 目標=90.0)
print(f"C線（目標90%）：{訊息}")
```

### 4-2　模組化設計——將生產報表系統拆成多個函數

每個函數只負責一件事，最後在主程式中組合使用。

```python
# ── 模組1：計算單條生產線的統計數字 ──
def 分析生產線(名稱, 總產量, 不良品數, 目標良率=95.0):
    """回傳一條生產線的分析結果（字典格式）"""
    良率 = 計算良率(總產量, 不良品數)
    達標, 狀態 = 判斷達標(良率, 目標良率)
    return {"名稱": 名稱, "總產量": 總產量, "良率": 良率, "達標": 達標, "狀態": 狀態}


# ── 模組2：列印單條生產線的資訊 ──
def 列印生產線結果(結果):
    print(
        f"  {結果['名稱']}：產量 {結果['總產量']} 件，"
        f"良率 {結果['良率']:.1f}%  {結果['狀態']}"
    )


# ── 模組3：產生完整報表 ──
def 產生日報表(生產資料):
    """輸入多條生產線資料，輸出完整日報"""
    print("=" * 45)
    print("           生產日報表")
    print("=" * 45)

    所有良率 = []
    未達標線 = []

    for 線名, 產量, 不良 in 生產資料:
        結果 = 分析生產線(線名, 產量, 不良)
        列印生產線結果(結果)
        所有良率.append(結果["良率"])
        if not 結果["達標"]:
            未達標線.append(結果["名稱"])

    print("-" * 45)
    平均良率 = sum(所有良率) / len(所有良率)
    print(f"  全廠平均良率：{平均良率:.1f}%")
    if 未達標線:
        print(f"  ⚠ 需要改善：{未達標線}")
    else:
        print("  ✓ 所有生產線均達標")
    print("=" * 45)


# ── 主程式：呼叫報表函數 ──
今日資料 = [
    ("A線", 500, 12),
    ("B線", 320, 8),
    ("C線", 410, 31),
    ("D線", 280, 4),
]

產生日報表(今日資料)
```

### 5-2　資料篩選與排序

```python
# 篩選（Filtering）：找出符合條件的資料

良率記錄 = [0.976, 0.981, 0.968, 0.943, 0.972, 0.985, 0.961, 0.978, 0.924, 0.969]

目標 = 0.95
未達標記錄 = []

for i, 良率 in enumerate(良率記錄):
    if 良率 < 目標:
        未達標記錄.append((i + 1, 良率))  # 儲存 (天數, 良率) 的配對

print("未達標日期：")
for 天數, 良率 in 未達標記錄:
    print(f"  第 {天數:2d} 天：{良率:.1%}")
print(f"未達標天數：{len(未達標記錄)} 天 / {len(良率記錄)} 天")
```

### 5-3　完整案例：感測器資料分析流程

```python
# 完整示範：讀取 → 清洗 → 分析 → 輸出

# ── 模擬讀取一天的溫度感測器資料（每小時一筆，共24筆）──
原始溫度資料 = [
    22.1,
    22.3,
    22.0,
    21.8,
    21.5,
    21.9,  # 0:00 ~ 5:00
    22.4,
    23.1,
    23.5,
    23.8,
    24.2,
    24.5,  # 6:00 ~ 11:00
    None,
    24.8,
    26.1,
    25.9,
    24.3,
    23.7,  # 12:00（離線）~ 17:00
    23.2,
    22.8,
    22.5,
    22.1,
    21.9,
    22.0,  # 18:00 ~ 23:00
]

上限 = 25.0
下限 = 20.0

# ── 步驟1：清洗資料（移除空值）──
有效資料 = [t for t in 原始溫度資料 if t is not None]
缺失筆數 = len(原始溫度資料) - len(有效資料)

# ── 步驟2：彙整統計 ──
平均溫度 = sum(有效資料) / len(有效資料)
最高溫度 = max(有效資料)
最低溫度 = min(有效資料)

# ── 步驟3：找出異常時段 ──
異常紀錄 = []
for i, 溫度 in enumerate(原始溫度資料):
    if 溫度 is None:
        continue
    if 溫度 > 上限 or 溫度 < 下限:
        異常紀錄.append((i, 溫度))

# ── 步驟4：輸出報告 ──
print("=" * 40)
print("  溫度感測器日分析報告")
print("=" * 40)
print(f"  資料筆數：{len(原始溫度資料)} 筆（缺失 {缺失筆數} 筆）")
print(f"  平均溫度：{平均溫度:.1f}°C")
print(f"  最高溫度：{最高溫度:.1f}°C")
print(f"  最低溫度：{最低溫度:.1f}°C")
print(f"  正常範圍：{下限}°C ~ {上限}°C")
print("-" * 40)
if 異常紀錄:
    print(f"  ⚠ 異常時段（共 {len(異常紀錄)} 筆）：")
    for 時間, 溫度 in 異常紀錄:
        print(f"     {時間:02d}:00  {溫度:.1f}°C  ← 超出範圍")
else:
    print("  ✓ 全天溫度均在正常範圍內")
print("=" * 40)
```
